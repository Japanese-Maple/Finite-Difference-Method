## Thermal Diffusion

The 2D Heat Equation is a parabolic partial differential equation that describes how the distribution of heat (or variation in temperature) evolves over time in a solid medium. It is given by:

$$\frac{\partial u}{\partial t} = \alpha \left( \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} \right)$$

Where:
* $u(x,y,t)$ is the temperature at spatial coordinates $(x,y)$ and time $t$.
* $\alpha$ is the **thermal diffusivity** of the material (how fast heat spreads).

To solve this numerically, our `Heat_Equation` class uses a **Finite Difference Method (FDM)** with an explicit time-stepping loop. To maintain stability, the solver internally enforces the Courant–Friedrichs–Lewy (CFL) condition constraint.

---

### Defining the Domain & Conditions:

```python
#                  1. System Parameters

N = 90             # Grid resolution (N x N)
H = 2.0            # Size of the spatial domain
alpha = 1.0        # Thermal diffusivity
time_s = 1500      # Number of time steps to simulate

# 2. Initial Condition 
#    (Cold plate)
def initial_condition(x, y):
    return np.zeros_like(x)

# 3. Boundary Conditions 
#    (Oscillating heat sources at the edges)

def upper_bc(x, t):
    return 13 * np.sin(3 * np.pi * x) * np.cos(50 * t)

def lower_bc(x, t):
    return 13 * np.sin(3 * np.pi * x) * np.cos(50 * t)

def right_bc(y, t):
    return np.reshape(13 * np.sin(3 * np.pi * y) * np.cos(50 * t), (N - 2, 1))

def left_bc(y, t):
    return np.reshape(13 * np.sin(3 * np.pi * y) * np.cos(50 * t), (N - 2, 1))
```

### Running the Simulation:

```python
# Paths

video_2d_target = os.path.join(output_dir, "Heat_Equation_Simulation_2D.mp4")
video_3d_target = os.path.join(output_dir, "Heat_Equation_Simulation_3D.mp4")
```

```python
HE = Heat_Equation(
    H=H, 
    n=N, 
    alpha=alpha, 
    simulation_time=time_s, 
    IC=initial_condition, 
    upper=upper_bc, 
    lower=lower_bc, 
    right=right_bc, 
    left=left_bc, 
    dpi=100,
    number_of_workers=15 # !!! Depends on your computer !!!
)

print("Solving the 2D Heat Equation...")
sol, vmax, vmin = HE.solve_Heat_Equation_4BC()
```

Saving the Animations:

```python
print("\nInitiating 2D Rendering Pipeline...")
HE.render_Heat_Eqn_2D(sol, vmax, vmin, video_filename=video_2d_target)

print("\nInitiating 3D Rendering Pipeline...")
HE.render_Heat_Eqn_3D(sol, vmax, vmin, video_filename=video_3d_target)
```
